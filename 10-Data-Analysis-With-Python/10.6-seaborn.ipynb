#### 1. Line Plot (with built-in style)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset('flights')

# Line plot showing passenger trends
sns.lineplot(data=data, x='year', y='passengers')

plt.title('Passenger Growth Over Time')
plt.show()
```

#### 2. Bar-plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset('titanic')

# Bar plot showing average age by passenger class
sns.barplot(data=data, x='class', y='age')

plt.title('Average Age by Passenger Class')
plt.show()
```

#### 3. Scatter Plot (with Regression Line) 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset('tips')

# Scatter plot with regression line
sns.regplot(data=data, x='total_bill', y='tip')

plt.title('Bill Amount vs Tip Amount')
plt.show()
```

```python
##create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
## Line Plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
## Categorical Plots
## BAr Plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plot of Total Bill By Day')
plt.show()
```

```python
## Box Plot
sns.boxplot(x="day",y='total_bill',data=tips)
```

```python
## Violin Plot

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
### Histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# Pairplot
sns.pairplot(tips)
```

```python
import pandas as pd
sales_df=pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
## Plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category',y="Total Revenue",data=sales_df,estimator=sum)
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
## Plot total sales by Region
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y="Total Revenue",data=sales_df,estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```
