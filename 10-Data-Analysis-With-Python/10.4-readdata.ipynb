- ðŸ’» Example 1: Parsing an HTML Document
- Here's a simple example of using lxml to parse HTML content:

```python
from lxml import html

# Sample HTML string
html_content = """
<html>
    <head><title>Sample Page</title></head>
    <body>
        <h1>Welcome to Python!</h1>
        <p class="intro">Learn data science with Python libraries like pandas and numpy.</p>
        <p>Visit our <a href="https://example.com">website</a>.</p>
    </body>
</html>
"""

# Parsing the HTML
tree = html.fromstring(html_content)

# Extracting data using XPath
title = tree.xpath('//title/text()')[0]  # Extracts <title> content
paragraphs = tree.xpath('//p/text()')     # Extracts all paragraph text
link = tree.xpath('//a/@href')[0]        # Extracts link URLs

print("Title:", title)
print("Paragraphs:", paragraphs)
print("Link:", link)
```

- ðŸ’» Example 2: Parsing XML Data
- lxml is highly efficient for XML parsing as well.

```python
from lxml import etree

# Sample XML string
xml_data = """
<library>
    <book id="1">
        <title>Python Crash Course</title>
        <author>Eric Matthes</author>
    </book>
    <book id="2">
        <title>Deep Learning with Python</title>
        <author>FranÃ§ois Chollet</author>
    </book>
</library>
"""

# Parsing XML data
root = etree.fromstring(xml_data)

# Extract book titles
titles = root.xpath('//book/title/text()')

print("Book Titles:", titles)
```

#### ðŸ”Ž html5lib and beautifulsoup4 Libraries Explained
- Both `html5lib` and `beautifulsoup4` are `popular Python libraries` primarily used `for web scraping and HTML/XML` parsing. 
- While they serve similar purposes, they have distinct functionalities. Here's a breakdown:

### 1. html5lib Library
ðŸ”¹ Purpose:
html5lib is an HTML parser that parses HTML documents exactly the way a web browser does â€” even with broken or poorly formatted HTML.

ðŸ”¹ Key Features:
- âœ… Parses HTML following the official HTML5 specification.
- âœ… Handles malformed HTML gracefully.
- âœ… Produces a clean and structured DOM tree (Document Object Model).
- âœ… Ensures maximum compatibility with web pages.

```python
from bs4 import BeautifulSoup

# Sample malformed HTML content
html_content = """
<html>
    <head><title>Sample Page</title></head>
    <body>
        <h1>Welcome to Web Scraping!</h1>
        <p>Python libraries like <b>BeautifulSoup</b> and <i>html5lib</i> are powerful.</p>
    </body>
</html>
"""

# Parsing with html5lib
soup = BeautifulSoup(html_content, 'html5lib')

# Extracting data
print(soup.title.text)  # Extracts <title> content
print(soup.h1.text)     # Extracts <h1> content
print(soup.p.text)      # Extracts <p> content
```

#### When to Use html5lib?
- âœ… Best for parsing poorly formatted or broken HTML.
- âœ… Ideal when dealing with web pages that aren't strictly valid HTML.

### 2. beautifulsoup4 Library
ðŸ”¹ `Purpose:`
`beautifulsoup4` (commonly referred to as BeautifulSoup) is a powerful library `used to extract data` from `HTML and XML` documents.

ðŸ”¹ `Key Features:`
- âœ… Provides a user-friendly API for navigating and searching the HTML/XML tree.
- âœ… Supports multiple parsers like:
- html.parser (built-in Python parser)
- lxml (faster parser for well-structured HTML/XML)
- html5lib (best for handling messy or broken HTML).
- âœ… Excellent for tasks like web scraping, data extraction, and content manipulation.

```python
from bs4 import BeautifulSoup
import requests

# Web scraping example
url = "https://example.com"
response = requests.get(url)

# Parsing with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Extracting content
title = soup.title.text
all_links = [a['href'] for a in soup.find_all('a') if 'href' in a.attrs]

print("Title:", title)
print("Links:", all_links)
```

#### When to Use beautifulsoup4?
- âœ… Best for general-purpose HTML/XML parsing.
- âœ… Perfect when combining with libraries like requests for web scraping.
- âœ… Great for navigating complex document trees using .find(), .find_all(), or .select() methods.

#### **3. Key Differences Between `html5lib` and `beautifulsoup4`**

| **Feature**               | **`html5lib`**            | **`beautifulsoup4`**             |
|---------------------------|----------------------------|----------------------------------|
| **Purpose**                | HTML5-compliant parser for fixing messy HTML. | Flexible parser for HTML/XML with various parsing options. |
| **Speed**                  | Slower due to strict compliance with HTML5 standards. | Faster when used with `lxml` parser. |
| **Handling Broken HTML**   | Excellent for parsing **malformed** or **incomplete** HTML. | Can handle malformed HTML, but not as effectively as `html5lib`. |
| **Parser Integration**     | Used as a **parser** with `BeautifulSoup`. | A complete library that can integrate with `html5lib`, `lxml`, or `html.parser`. |
| **Best Use Case**          | Parsing highly complex or messy web pages. | Ideal for general-purpose web scraping and data extraction. |



#### 4. Which One Should You Use?
- âœ… Use beautifulsoup4 with html.parser (faster) or lxml (faster with clean HTML).
- âœ… Use beautifulsoup4 with html5lib when dealing with poorly structured HTML or websites with complex tags.

ðŸ”¹ Recommended Combination for Web Scraping:

```python
from bs4 import BeautifulSoup
import requests

url = "https://example.com"
response = requests.get(url)

# Best combination for flexibility and speed
soup = BeautifulSoup(response.text, 'lxml')  

print(soup.title.text)
```

#### 5. Summary
- `html5lib` â†’ Best for handling broken HTML or messy web pages.
- `beautifulsoup4` â†’ Excellent for data extraction, web scraping, and navigating HTML trees.
- `Combine` `beautifulsoup4 with html5lib` or `lxml` for `optimal results` based on your data structure.

### other examples using this library

#### 1. Writing Data to Excel

```python
from openpyxl import Workbook

# Create a workbook and select the active worksheet
wb = Workbook()
ws = wb.active

# Add data to cells
ws['A1'] = "Name"
ws['B1'] = "Age"
ws.append(["Alice", 30])
ws.append(["Bob", 25])

# Save the workbook
wb.save("sample.xlsx")
```

#### 2. Reading Data from Excel

```python
from openpyxl import load_workbook

# Load the workbook
wb = load_workbook("sample.xlsx")
ws = wb.active

# Access cell data
print(ws['A1'].value)   # Output: Name
print(ws['B2'].value)   # Output: 30

# Loop through rows
for row in ws.iter_rows(values_only=True):
    print(row)
```

#### 3. Modifying Excel Data

```python
ws['B2'] = 32  # Change Alice's age to 32
wb.save("sample.xlsx")
```
